# To verify the assumption of multicollinearity using Variance Inflation factor!!

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X.shape,Y.shape)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
vif_chd = pd.DataFrame()
vif_chd['column']=X.columns[0:11]
vif_chd['vif']= [variance_inflation_factor(X,i) for i in range(X_train.shape[1])]
vif_chd.sort_values(by='vif',ascending=False)

```

# What to do: Here, we will remove variables one by one accordingly the VIF values.

# **Criteria behind this:**
# The variable for which the VIF value greater than 5, it is not considered in the Logistic regression.So, at the first step, we will run logistic regression by taking all the variables and then after remove "BMI".

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()
lr_model = lr.fit(X_train,Y_train)
```

```python
Y_pred = lr.predict(X_test)
Y_pred
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
print(classification_report(Y_test,Y_pred))
```

```python
# Here, 74% variability in "Outcome" is explained by the Logistic Regression model.
```
